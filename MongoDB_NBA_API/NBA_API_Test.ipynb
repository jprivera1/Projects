```python
from nba_api.stats.static import teams
from nba_api.stats.static import players
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def one_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}
    for dict_ in list_dict:
        for key, value in dict_.items():
            out_dict[key].append(value)
    return out_dict    
```

## GAME FINDER

```python
from nba_api.stats.endpoints import leaguegamefinder
```

```python
#Get the ID of a specific team (in this case Miami Heat) just from its abbreviation
MIA = [team for team in nba_teams if team['abbreviation'] == 'MIA'][0]
MIA_Id = MIA['id']
```

```python
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=MIA_Id) #team = MIA
gamefinder.get_json()
```

```python
games = gamefinder.get_data_frames()[0]
games.head()
```

As you can see above, the season ID is 5 digits. I believe the last 4 will always be the current season (2018 for the 2018-19 season). We can do a sanity check and look at how many games the Celtics have played in recent years.

```python
games.groupby(games.SEASON_ID.str[-4:])[['GAME_ID']].count().loc['2015':]
```

### Filter to Games in the 2020-21 Season


```python
# Subset the games to when the last 4 digits of SEASON_ID were 2020.
games_2021 = games[games.SEASON_ID.str[-4:] == '2020']
games_2021.head()
```
